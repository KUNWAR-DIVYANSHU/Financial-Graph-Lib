```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../ACC.csv")
df.head(10)
```

# Plotting Graph
we have successfully fetch the data with 15 columns . Let's plot a area graph where date is plotted on x-axis while close price is plotted on y-axis.

First let's remove date as the default index and change the format of date to date-month-year to do so you need to use pd.to_datetime() function and then format the string.

```python
df = df.reset_index()
df['date'] = pd.to_datetime(df['date']).dt.strftime('%d-%b-%y')
```

#### create a subplot with variable name fig and it's axes as ax .
Plot index vs close price and then make the graph looks beautiful by adding labels and titles. To prevent messing up text on x-axis. you can use [::150] to select every 150th index from the list.

```python
fig, ax = plt.subplots(1, figsize=(18,8))
plt.fill_between(df.index,df['close'], color='green')
ax.set_ylabel('Closing Price')
ax.set_title('ACC')
ax.set_xticks(df.index[::300])
ax.set_xticklabels(df["date"][::300])
"graph plotted"
```
